```python
growth_factors = [3.20, 1.15, 4.80, 1.30, 1.12, 1.85, 1.75, 1.28]

initial_porgs = 5

#the product of growth factors
product_of_factors = 1
for factor in growth_factors:
    product_of_factors *= factor

# the total number of Porgs after eight hours
total_porgs = initial_porgs * product_of_factors


print("Total Porgs after eight hours:", total_porgs)
```

```python
import numpy as np


growth_rates = [220, 15, 380, 30, 12, 85, 75, 28]

# Calculate arithmetic mean
arithmetic_mean = np.mean(growth_rates)

# Calculate geometric mean
# Convert growth rates to growth factors
growth_factors = [(1 + rate/100) for rate in growth_rates]

# Calculate the product of growth factors
product_of_factors = np.prod(growth_factors)

# Calculate the geometric mean
geometric_mean = product_of_factors ** (1/len(growth_rates))


# Estimating Porgs using means
estimated_porgs_arithmetic = initial_porgs * (1 + arithmetic_mean/100) ** 8
estimated_porgs_geometric = initial_porgs * geometric_mean ** 8


print("Arithmetic Mean of Growth Rates:", arithmetic_mean)
print("Geometric Mean of Growth Rates:", geometric_mean)
print("Estimated Porgs after eight intervals (Arithmetic Mean):", estimated_porgs_arithmetic)
print("Estimated Porgs after eight intervals (Geometric Mean):", estimated_porgs_geometric)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given growth factors
growth_factors = [3.20, 1.15, 4.80, 1.30, 1.12, 1.85, 1.75, 1.28]

# Initial Porgs
initial_porgs = 5

# Calculate the product of growth factors
product_of_factors = np.prod(growth_factors)

# Calculate the number of Porgs at each hour
# Note that we used '+1' because the indexing begins at 0
porgs_at_each_hour = [initial_porgs * product_of_factors ** (i+1) for i in range(len(growth_factors))]

# Create a bar plot
hours = ['4:00am', '5:00am', '6:00am', '7:00am', '8:00am', '9:00am', '10:00am', '11:00am']
plt.bar(hours, porgs_at_each_hour, color='skyblue')
plt.xlabel('Hour')
plt.ylabel('Number of Porgs')
plt.title('Number of Porgs at Each Hour')
plt.show()
```
